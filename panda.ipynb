```python
print("Let's create a recipe!")

# Get recipe name
recipe_name = input("Enter the name of the recipe: ")

# Get ingredients
ingredients = []
print("\nEnter ingredients one by one. Type 'done' when finished.")
while True:
    ingredient = input(f"Ingredient {len(ingredients) + 1}: ")
    if ingredient.lower() == 'done':
        break
    if ingredient:
        ingredients.append(ingredient)

# Get preparation steps
steps = []
print("\nEnter preparation steps one by one. Type 'done' when finished.")
while True:
    step = input(f"Step {len(steps) + 1}: ")
    if step.lower() == 'done':
        break
    if step:
        steps.append(step)

# Display the complete recipe
print("\n--- Your Recipe ---")
print(f"Recipe Name: {recipe_name}")

print("\nIngredients:")
if ingredients:
    for i, ingredient in enumerate(ingredients):
        print(f"{i + 1}. {ingredient}")
else:
    print("No ingredients added.")

print("\nPreparation Steps:")
if steps:
    for i, step in enumerate(steps):
        print(f"{i + 1}. {step}")
else:
    print("No steps added.")

print("-------------------")
```
