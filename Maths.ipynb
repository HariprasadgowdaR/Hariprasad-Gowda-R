```python
import math
import matplotlib.pyplot as plt
import numpy as np

rainfall_list=[]
fertilizer_list=[]

print("Enter 10 rainfall values in mm:")
for i in range(10):
  val=float(input(f"Rainfall {i+1}:"))
  rainfall_list.append(val)

print("\n Enter 10 fertilizer values in units")
for i in range(10):
  val=float(input(f"Fertilizer {i+1}:"))
  fertilizer_list.append(val)

def predict_yield(rainfall,fertilizer,base_yield=2000):
  rainfall_co_efficient=10
  fertilizer_co_efficient=15
  predicted=base_yield+(rainfall*rainfall_co_efficient)+(fertilizer*fertilizer_co_efficient)
  return predicted

yield_list=[predict_yield(r,f) for r,f in zip(rainfall_list,fertilizer_list)]

def profit_function(fertilizer_amount,price_per_kg=20,fertilizer_cost_per_unit=50):
  yield_increase=500*math.sqrt(fertilizer_amount)
  revenue=yield_increase*price_per_kg
  cost=fertilizer_amount*fertilizer_cost_per_unit
  profit=revenue-cost
  return profit

profit_list=[profit_function(f) for f in fertilizer_list]

print("\n Results displayed\n")
for i in range(10):
  print(f"Rainfall: {rainfall_list[i]}mm , Fertlizer: {fertilizer_list[i]} units-----> Yield: {yield_list[i]:.2f} kg/ha , Profit: Rs.{profit_list[i]:.2f}")
```
