```python
import pandas as pd
df=pd.read_csv('data.csv.xls')
print(df.describe())
```

```python
import pandas as pd
mean_duration = df['Duration'].mean()
mean_calories = df['Calories'].mean()
print(f"The average duration is: {mean_duration:.3f} minutes")
print(f"The average calories burned is: {mean_calories:.3f} Calories")

high_duration_df = df[(df['Duration'] > mean_duration) & (df['Calories'] > mean_calories)]
print("High duration of metabolism (Duration > average duration AND Calories > average calories):")
display(high_duration_df.head())


print(f"Shape of the original DataFrame: {df.shape}")
print(f"Shape of the high duration & metabolism DataFrame: {high_duration_df.shape}")
```

```python
import pandas as pd
avg_duration=df['Duration'].mean()
avg_calories=df['Calories'].mean()
print(f"The mean durartion of metabolism:{avg_duration:4f} mins")
print(f"The average calories burnt:{avg_calories:4f} cal")

high_meatabolism=df[(df['Duration']>avg_duration) & (df['Calories']>avg_calories)]
print(f"High metabolism is seen in: \n{high_meatabolism.to_string()}")
```
