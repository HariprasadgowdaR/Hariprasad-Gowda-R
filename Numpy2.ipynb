```python
import numpy as np
crop_yield=np.array([
    [10,10.2,11,12,12.3],
    [12,11.2,16.5,12,12.3],
    [13,15.2,11.8,12,12.3],
    [14,14.2,19,12,12.3]
])

print("Crop Yield Array (tons:)\n",crop_yield)
print(f'\nShape:{crop_yield.shape}')
print(f'Data Type:{crop_yield.dtype}')
print(f'Total elements:{crop_yield.size}')
print(f'Number of Dimensions"{crop_yield.ndim}')

total_yield=np.sum(crop_yield)
avg_yield_per_farm=np.mean(crop_yield,axis=1)
avg_yield_season=np.mean(crop_yield,axis=0)
max_yield=np.max(crop_yield)
min_yield=np.min(crop_yield)

print(total_yield)
print(avg_yield_per_farm)
print(avg_yield_season)
print(max_yield)
print(min_yield)
```
