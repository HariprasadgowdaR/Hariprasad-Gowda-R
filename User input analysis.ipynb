```python
import numpy as np
while True:
    try:
        n = int(input("Enter the number of months: "))
        if n <= 0:
            print("Please enter a positive number.")
        else:
            break
    except ValueError:
        print("Invalid input. Please enter a whole number.")

print("Enter monthly rainfall data:")
rainfall_data = []
for i in range(n):
    while True:
        try:
            rain = int(input(f"Enter rainfall for month {i+1} (mm): "))
            rainfall_data.append(rain)
            break
        except ValueError:
            print("Invalid input. Please enter a whole number.")
print("New monthly rainfall data entered:", rainfall_data)

print("Enter monthly yield data:")
new_yield_data = []
for i in range(n):
    while True:
        try:
            yield_val = float(input(f"Enter yield for month {i+1} (tons/ha): "))
            new_yield_data.append(yield_val)
            break
        except ValueError:
            print("Invalid input. Please enter a number.")
print("New monthly yield data entered:", new_yield_data)

print("RAINFALL STATISTICS")
print(f'Mean: {np.mean(rainfall_data):.2f} mm')
print(f'Median:{np.median(rainfall_data):.2f} mm')
print(f'Std Deviation:{np.std(rainfall_data):.2f} mm')
print(f'Varience:{np.var(rainfall_data):.2f} mm')
print(f'Minimum: {np.min(rainfall_data):.2f} mm (Month{np.argmin(rainfall_data)+1})')
print(f'Maximum: {np.max(rainfall_data):.2f} mm (Month{np.argmax(rainfall_data)+1})')
print(f'Range: {np.ptp(rainfall_data): .2f} mm')

print("\nYIELD STATISTICS")
print(f'Mean:{np.mean(new_yield_data):.2f} tons/ha')
print(f'Median:{np.median(new_yield_data):.2f} tons/ha')
print(f'Std Deviation:{np.std(new_yield_data):.2f} tons/ha')

print("\nPERCENILES")
percentile=[12,50,75,100]
rainfall_percentile=np.percentile(rainfall_data,percentile)
for p, val in zip(percentile,rainfall_percentile):
    print(f'{p}th percentile rainfall: {val:.2f} mm')

print("\nCORRELATION ANALYSIS")
correlation_matrix=np.corrcoef(rainfall_data,new_yield_data)
correlation=correlation_matrix[0,1]
print('Interception: Strong Positive Correlation' if correlation>0.7 else ('Moderate Correlation' if correlation>0.4 else 'Weak correlation'))

print("\nCUMULATIVE RAINFALL")
cumsum=np.cumsum(rainfall_data)
print('Month\trainfall\tcumulative')
for month,(rain,cum) in enumerate (zip(rainfall_data,cumsum),1):
  print(f'{month}\t{rain}\t{cum}')


import matplotlib.pyplot as plt

months = range(1, len(cumsum)+1) # Update months to match the length of rainfall data
plt.figure(figsize=(10, 6))
plt.plot(months, cumsum, marker='o', linestyle='--')
plt.title('Cumulative Rainfall Over Months')
plt.xlabel('Month')
plt.ylabel('Cumulative Rainfall (mm)')
plt.xticks(months)
plt.grid(True)
plt.show()

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(x=rainfall_data, y=new_yield_data)
plt.title('Rainfall vs. Crop Yield')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Yields (tons/ha)')
plt.show()
```
