```python
import numpy as np
rainfall=([25,45,52,68,120,180,220,195,160,110,75,58])
yields=np.array([2.1,2.3,3.8,4.2,5.8,6.5,5.9,5.2,3.8,2.9,2.4,2])
print("RAINFALL STATISTICS")
print(f'Mean: {np.mean(rainfall):.2f} mm')
print(f'Median:{np.median(rainfall):.2f} mm')
print(f'Std Deviation:{np.std(rainfall):.2f} mm')
print(f'Varience:{np.var(rainfall):.2f} mm')
print(f'Minimum: {np.min(rainfall):.2f} mm (Month{np.argmin(rainfall)+1})')
print(f'Maximum: {np.max(rainfall):.2f} mm (Month{np.argmax(rainfall)+1})')
print(f'Range: {np.ptp(rainfall): .2f} mm')

print("\nYIELD STATISTICS")
print(f'Mean:{np.mean(yields):.2f} tons/ha')
print(f'Median:{np.median(yields):.2f} tons/ha')
print(f'Std Deviation:{np.std(yields):.2f} tons/ha')

print("\nPERCENILES")
percentile=[12,50,75,100]
rainfall_percentile=np.percentile(rainfall,percentile)
for p, val in zip(percentile,rainfall_percentile):
    print(f'{p}th percentile rainfall: {val:.2f} mm')

print("\nCORRELATION ANALYSIS")
correlation_matrix=np.corrcoef(rainfall,yields)
correlation=correlation_matrix[0,1]
print('Interception: Strong Positive Correlation' if correlation>0.7 else ('Moderate Correlation' if correlation>0.4 else 'Weak correlation'))

print("\nCUMULATIVE RAINFALL")
cumsum=np.cumsum(rainfall)
print('Month\trainfall\tcumulative')
for month,(rain,cum) in enumerate (zip(rainfall,cumsum),1):
  print(f'{month}\t{rain}\t{cum}')


import matplotlib.pyplot as plt

months = range(1, 13)
plt.figure(figsize=(10, 6))
plt.plot(months, cumsum, marker='o', linestyle='--')
plt.title('Cumulative Rainfall Over Months')
plt.xlabel('Month')
plt.ylabel('Cumulative Rainfall (mm)')
plt.xticks(months)
plt.grid(True)
plt.show()

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(x=rainfall, y=yields)
plt.title('Rainfall vs. Crop Yield')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Yields (tons/ha)')
plt.show()
```
