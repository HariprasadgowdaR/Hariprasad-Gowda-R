```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,25,20,30,40]
plt.plot(x,y,marker='o',
linestyle='--',color='blue',
linewidth=2,markersize=8,
label='sales')
plt.title("monthly sales trends")
plt.xlabel("month")
plt.ylabel("sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(50)
y=np.random.rand(50)
sizes=1000*np.random.rand(50)
colors=np.random.rand(50)
plt.scatter(x,y ,s=sizes,c=colors,cmap='viridis',alpha=0.6)
plt.colorbar()
plt.title("advanced scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
