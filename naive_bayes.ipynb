```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB

data={
    'weather':['Sunny','Sunny','Rainy','Rainy','Cloudy','Sunny','cloudy','Rainy'],
    'Soil':['Dry','Wet','Wet','Dry','Dry','Dry','Wet','Wet'],
    'temperature':['Hot','Hot','Cool','Cool','Cool','Hot','Hot', 'Hot'],
    'irrigate':['Yes','No','No','Yes','Yes','Yes','No','No']
}
df=pd.DataFrame(data)
print("dataset:\n",df)

Le_weather=LabelEncoder()
Le_soil=LabelEncoder()
Le_temperature=LabelEncoder()
Le_irrigate=LabelEncoder()
df['weather']=Le_weather.fit_transform(df['weather'])
df['Soil']=Le_soil.fit_transform(df['Soil'])
df['temperature']=Le_temperature.fit_transform(df['temperature'])
df['irrigate']=Le_irrigate.fit_transform(df['irrigate'])
print("\nEncoded Dataset:\n",df)

X=df[['weather','Soil','temperature']]
y=df['irrigate']

model=CategoricalNB()
model.fit(X,y)

test_data=pd.DataFrame({
    'weather':['Sunny'],
    'Soil':['Dry'],
    'temperature':['Hot']
})
test_data['weather']=Le_weather.transform(test_data['weather'])
test_data['Soil']=Le_soil.transform(test_data['Soil'])
test_data['temperature']=Le_temperature.transform(test_data['temperature'])

predication=model.predict(test_data)
result=Le_irrigate.inverse_transform(predication)
print("\nTest input:Sunny,Dry,Hot")
print("predition",result[0])

prob=model.predict_proba(test_data)
print("Class Probablities",prob)
```
