```python
import numpy as np

A = np.random.randint(1, 100, size=(5, 5))
B = np.random.randint(1, 100, size=(5, 5))

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)

dotpro = A @ B
print("\n Dot Product:\n", dotpro)

Trans = A.T
print("\n Transpose of A :\n", Trans)

multi = A * B
print("\n Multiplication (A * B):\n", multi)

det_A = np.linalg.det(A)
print("\n Determinant of A :", det_A)

if det_A != 0:
    A_inv = np.linalg.inv(A)
    print("\n Inverse of A \n", A_inv)
else:
    print("\n Inverse of A: Determinant is zero, inverse does not exist.")

rank_A = np.linalg.matrix_rank(A)
print("\n Rank of A :", rank_A)


```

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
print("\n Eigenvalues of A :\n", eigenvalues)
print("\nEigenvectors of A :\n", eigenvectors)
```
